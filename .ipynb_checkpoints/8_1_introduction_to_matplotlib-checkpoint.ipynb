```python
# 1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
#2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1]) # add axes
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 500]) # add some data
plt.show()
type(fig), type(ax)
```

## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in jupyter
%matplotlib module://matplotlib_inline.backend_inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot (Object-Oriented Method)
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
       ylabel='y-axis')

# 5. Save & show (you save the whole figure)
fig.savefig('./plots/1_sample-plot.png')
plt.show()
```

## Making figures with Numpy arrays

We want:

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
# create some data
x = np.linspace(1, 10, 100)
x[:10]
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
plt.show()
```

```python
# Use same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Make a plot from dictionary

nut_butter_prices = {'Almond butter': 10,
                    'Peanut butter': 8,
                    'Chashew butter': 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title='Dan\'s nut butter store', 
       ylabel='Price ($)');
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values())
```

```python
# Make some data
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```
