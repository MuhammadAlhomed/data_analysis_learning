# Customizing Matplotlib Plots & Getting stylish

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Get datasets

car_sales = pd.read_csv('data/car-sales.csv')
# fix the Price column from string to int
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$\,]','', regex=True).replace(r'\.\d*','', regex=True).astype(int)


heart_disease = pd.read_csv('data/heart-disease.csv')

over_50 = heart_disease[heart_disease['age'] > 50] # Filtered to be over 50
```

Previous Plot

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               figsize=(10,10),
                                sharex=True)

# Add data to ax0
scatter = ax0.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target']
)

# Customize ax0
ax0.set(
    title='Heart Disease and Cholestrol Levels',
    ylabel='Cholestrol'
)

# Add a legent to ax0
ax0.legend(*scatter.legend_elements(), title='Target')
# Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

########################################################

# Add data to ax1
scatter = ax1.scatter(
    x=over_50['age'],
    y=over_50['thalach'],
    c=over_50['target']
)

# Customize ax1
ax1.set(
    title='Heart Disease and Max Heart Rate',
    xlabel='Age',
    ylabel='Max Heart Rate'
)

# Add a legent to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

#########################################################

# Add title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```

## Set the stile within a style

```python
plt.style.use('seaborn-v0_8-whitegrid')


'------------------------------ V REUSED CODE FROM BEFORE V ----------------------------'
# OO method From scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter' # <=========================== CHANGE HERE
                    )

# Customize the plot
ax.set(
    title='Heart Disease and Cholestrol Levels',
    xlabel='Age',
    ylabel='Cholestrol'
)

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
# Customizing x and y axis limits

'------------------------------ V REUSED CODE FROM BEFORE V ----------------------------'
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               figsize=(10,10),
                                sharex=True)

# Add data to ax0
scatter = ax0.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
    cmap='winter'
)

# Customize ax0
ax0.set(
    title='Heart Disease and Cholestrol Levels',
    ylabel='Cholestrol'
)

# Add a legent to ax0
ax0.legend(*scatter.legend_elements(), title='Target')

# Change x Limit
ax0.set_xlim((50,80)) # <======================================================= CHANGE HERE

# Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

########################################################

# Add data to ax1
scatter = ax1.scatter(
    x=over_50['age'],
    y=over_50['thalach'],
    c=over_50['target'],
    cmap='winter'
)

# Customize ax1
ax1.set(
    title='Heart Disease and Max Heart Rate',
    xlabel='Age',
    ylabel='Max Heart Rate'
)

# Add a legent to ax1
ax1.legend(*scatter.legend_elements(), title='Target')

# Change Y Limit
ax1.set_ylim((60,200)) # <======================================================= CHANGE HERE

# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

#########################################################

# Add title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
