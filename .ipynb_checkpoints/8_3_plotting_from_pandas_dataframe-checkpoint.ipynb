## Which plot should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use OO method.


```python
over_50 = heart_disease[heart_disease['age'] > 50] # Filtered to be over 50
over_50
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target' # Color it by the target column
)
```

```python
# OO Method mixed with pyplot method (passing DataFrame Plot data to 'ax')
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax # use the ax we created to plot it in.
            )

# Anything below is entirely exclusive to OO method, this is the real advantage of OO.
ax.set_xlim([45,100])
```

```python
# OO method From scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# Customize the plot
ax.set(
    title='Heart Disease and Cholestrol Levels',
    xlabel='Age',
    ylabel='Cholestrol'
)

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target')

# Add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--')
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               figsize=(10,10),
                                sharex=True)

# Add data to ax0
scatter = ax0.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target']
)

# Customize ax0
ax0.set(
    title='Heart Disease and Cholestrol Levels',
    ylabel='Cholestrol'
)

# Add a legent to ax0
ax0.legend(*scatter.legend_elements(), title='Target')
# Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

########################################################

# Add data to ax1
scatter = ax1.scatter(
    x=over_50['age'],
    y=over_50['thalach'],
    c=over_50['target']
)

# Customize ax1
ax1.set(
    title='Heart Disease and Max Heart Rate',
    xlabel='Age',
    ylabel='Max Heart Rate'
)

# Add a legent to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

```
